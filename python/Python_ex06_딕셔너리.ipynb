# 딕셔너리 

## 목표
- 딕셔너리에 대해 알 수 있다.
- 다양한 방법으로 딕셔너리를 생성하고 활용할 수 있다.

### 딕셔너리란?



```
dictionary = {Key : Value, Key : Value, ...}
```

- key : value 한쌍으로 이루어진 자료구조
- 순서가 없는 형태 (인덱스 번호 X)
- {}를 통하여 생성
- key : 변하지 않는 `유일한` 값 > 중복해서 사용시 덮어쓰기 됨.

- 자료구조의 필요성
  - 리스트 : 변화해도 되는 데이터를 저장할 때
  - 튜플 : 원본데이터를 저장할 때
  - 딕셔너리 : 순서가 필요없는 정보들을 저장할 때

````python
student = {"이름" : "최석진", "나이" : "21", "전화번호" : "010-0000-0000"}
student
````

````python
type(student), print(type(student))
````

````python
#key 값을 활용해서 val 추출
````

- key 값을 활용해서 val 추출 > 딕셔너리명[key]
- value 값은 변경이 가능 > 추가, 수정, 삭제 가능

````python
student["이름"]
````

````python
# 값 추가 > 추가할 값에 대한 새로운 key값을 대괄호 안에 넣어주고 대입
student["성별"] = "M"
student
````

````python
# value 수정 (나이 변경)
student["나이"] = 0
student
````

````python
# 값 삭제 > del 딕셔너리명["key"]
del student["성별"]
student
````

- 한개의 key에 여러개의 value 대응가능 > list 형태로 추가

````python
student["취미"] = ["코딩","여행","음악 감상"]
student
````

- value 색인방법 1 `대괄호[ ]`

````python
student['이름']
````

- value 색인방법 2 `dictionary.get(key)`

````python
student.get("이름")
````

- 두 방법의 차이는? > get함수는 없는 key값을 제공해도 오류 X > 예외처리 가능!!!

````python
student['생일']
````

````python
print(student.get("생일"))
````

- key값 확인

````python
student.keys()
````

````python
type(student.keys())
````

````python
list(student.keys()) # 강제 형 변환
````

- value값 확인

````python
student.values()
````

````python
type(student.values())
````

````python
for key in student.keys():
    print(key)
````

````python
for value in student.values():
    print(value)
````

- 전체 데이터를 리스트 형식으로 가져오기
  - ```dictionary.items()```

````python
student.items()
````

````python
for data in student.items():
    for i in data:
        print(i)
````

````python
for key, value in student.items():
    print(key,value)
````

- 딕셔너리 멤버연산자 활용
  - key in 딕셔너리명 : 멤버연산자는 딕셔너리의 key에만 동작

````python
'이름' in student #key 값만 활용 가능
````

````python
'최석진' in student #value 값은 활용 불가
````

````python
if '성별' in student:
    del student['성별']
else:
    print("해당 key값이 없습니다")
````

````python
print(student)
````
