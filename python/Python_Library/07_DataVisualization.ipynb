#### pandas의 plot을 활용한 시각화
- 연도별 입사자 추세 확인

````python
hr_info = pd.read_csv("./data/hr-info.csv",encoding = 'euc-kr')
````

````python
hr_info[['HIRE_DATE']]
````

- 입사연월일 데이터 (문자열) > 날짜타입 변경 > 연/월/일 을 각각 추출 가능
  ```
  pd.to_datetime()
  ```

````python
hr_info['HIRE_DATE'] = pd.to_datetime(hr_info['HIRE_DATE'])
````

````python
hr_info.info()
````

````python
hr_info['HIRE_DATE'].dt.year
````

- 새로운 컬럼 "HIRE_YEAR" 컬럼 추가 후 연도 데이터

````python
hr_info["HIRE_YEAR"] = hr_info['HIRE_DATE'].dt.year
hr_info
````

- 연도별 입사자수를 그래프로 그리려면 'value_counts' > 연도순으로 정렬

```
sort.index()
```
를 통해서 인덱스를 정렬해줌

````python
hr_info[['HIRE_YEAR']].value_counts().sort_index()
````

````python
hr_info[['HIRE_YEAR']].value_counts(ascending = True) # 이건 값을 내림차순으로 정렬
````

- `hr_info[['HIRE_YEAR']].value_counts(ascending = True)` : 값을 내림차순으로 정렬
- `hr_info[['HIRE_YEAR']].value_counts().sort_index()` : 인덱스를 정렬

- 데이터.plot() : pandas에서 제공하는 시각화

````python
hire_year_rs = hr_info['HIRE_YEAR'].value_counts().sort_index()
````

````python
hire_year_rs.plot(ylabel = 'HIRE_COUNT',grid = True) #
````

- ylabel 을 통해 y축의 이름을 설정할 수 있음
- grid를 통해 격자무늬 추가

````python
hire_year_rs.plot(ylabel = 'HIRE_COUNT', kind = 'barh') #
````

- 학위별 평균 연봉을 비교하는 시각화 (막대그래프 출력)

- 그래프의 한글 출력을 위해 인코딩 필요

````python
from matplotlib import rc
rc('font',family = 'Malgun Gothic')
````

````python
mean_salary = hr_info[['EDU_LEVEL','SALARY']].groupby('EDU_LEVEL').mean().round(1)
````

````python
mean_salary = mean_salary.loc[['고등학교 졸업','전문대 학위','학사 학위','석사 학위','박사 학위']]
# 학위 순서로 정렬
````

````python
mean_salary.plot(kind = 'bar',ylabel = '평균연봉',color = 'lightblue')
````

- 카테고리별 데이터는 bar차트로 그려주는 게 좋음

- groupby 함수안에 들어가야할 항목은 우리가 구별하고 싶은 기준을 넣어줘야함!!!!
- 데이터분석을 할 때, 가설 설정 : 학위가 올라가면 올라갈 수록 연봉이 높아질 것이다..

##### 학위, 성별을 고려한 평균연봉 시각화 

````python
hr_info.groupby('SALARY')[['GENDER','EDU_LEVEL']].mean()
````

> 에러 발생 : groupby() 사용시 기준이 될 데이터도 넣어줘야한다...

````python
hr_info[['GENDER','EDU_LEVEL','SALARY']].groupby(['GENDER','EDU_LEVEL']).mean().round(1)
````

- 기준이 2개 이므로 groupby

````python
hr_info[['GENDER','EDU_LEVEL','SALARY']].groupby(['EDU_LEVEL','GENDER']).mean().plot(kind='bar',color = 'lightpink')
````

- 같은 카테고리별로 비교할 수 있도록 (고등학교 졸업 - 남/녀) / (전문대 - 남/녀) > 이런식으로 보고 싶음 > 'pivot table'을 활용

- 데이터 재구조화 > `pivot table`

인덱스 (EDU_LEVEL)



````python
temp = hr_info[['GENDER','EDU_LEVEL','SALARY']].groupby(['EDU_LEVEL','GENDER']).mean().round(1)
temp
````

````python
temp.reset_index()
````

````python
temp = hr_info[['GENDER','EDU_LEVEL','SALARY']].groupby(['EDU_LEVEL','GENDER']).mean().round(1)
edu_gen_mean = temp.reset_index().pivot(index = 'EDU_LEVEL',columns = 'GENDER',values = 'SALARY')
edu_gen_mean
````

- EDU_LEVEL: index / GENDER : column / SALARY : values

````python
edu_gen_mean.plot(kind = 'bar',color = ['lightblue','hotpink'])
````

````python
eg_sort = edu_gen_mean.loc[['박사 학위', '석사 학위','학사 학위','전문대 학위','고등학교 졸업']]
eg_sort.plot(kind='barh',subplots = True,color = ['skyblue','hotpink'])
# male, female을 분리해서 보고 싶으
````

````python
eg_sort = edu_gen_mean.loc[['박사 학위', '석사 학위','학사 학위','전문대 학위','고등학교 졸업']]
eg_sort.plot(kind='barh',subplots = True,color = ['skyblue','hotpink'],xlim = (3000,9000))
# 범위 제한 xlim,ylim 활용해서 필요한 부분만 
````

````python
eg_sort.plot(kind='barh',subplots = True,color = ['skyblue','hotpink'],xlim = (3000,9000))
````

````python
ax = eg_sort.plot(kind='bar',subplots = True,color = ['skyblue','hotpink'],xlim = (3000,9000),rot = 0)
````

````python
axes = eg_sort.plot(
    kind='barh',
    subplots=True,
    color=['pink', 'skyblue'],
    xlim=(3000, 9000)
)

for ax in axes:
    ax.set_ylabel('EDU_LEVEL', rotation=0, labelpad = 20)
````

- 파이차트 > 비율 (전체 몇 퍼센트 정도 인지)

````python
hr_info['EDU_LEVEL'].value_counts().plot(kind = 'pie',
                                         autopct ='%.1f%%',
                                         title = '학위별 비율',
                                         colormap = 'Pastel1',
                                        ylabel = "",
                                        counterclock = False)
plt.show()
````

#### matplotlib을 활용한 시각화

```
plt.plot(데이터셋)
```

hire_year_rs를 데이터 프레임으로 받아오면 그림이 그려지지 않음 (Series형태로 받아야함)
```
hire_year_rs = hr_info[['HIRE_YEAR']].value_counts().sort_index()
```

````python
plt.plot(hire_year_rs)
plt.show()
````

````python
hire_year_rs.index # 리스트 형태로 나옴
````

````python
plt.figure(figsize=(15,8)) # 밑바탕의 사이즈 (가로,세로) > 처음에 크기를 먼저 정해주어야 함!
plt.plot(hire_year_rs,
        ls = '-', # line타입
        lw = 2, # 두께설정
        color = "#eb0788", # 색상 설정 : hex color
        marker = 'p', # 마커 표시}
        mfc = '#1d0af2', # 마커 색상
        mec = '#1d0af2', # 마커 테두리 색상
        ms = 3, # 마커 사이즈
        mew = 2, # 마커 테두리 두께
        label = '인원수')
plt.grid()
plt.xticks(hire_year_rs.index) # grid를 xticks값의 인덱스로 전달하여 grid를 데이터 위에 그려지도록 바꿈
plt.xlabel('입사년도')
plt.ylabel('인원수',labelpad = 20).set_rotation(0) # labelpad : 그래프로부터 제목 떨어뜨리기, set_rotation : 제목 돌리기
plt.title("연도별 입사인원 수", pad = 10) #pad를 통해 그래프에서 제목 떨어뜨리기
plt.legend() # 범례표시 > 위에 label에 명시해줘야함!
plt.show()
````

- 하나의 차트에 여러개의 plot 그리기

````python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]
plt.plot(x, label = 'x')
plt.plot(y, label = 'y')
plt.plot(z, label= 'z')
plt.legend() # 범례표시의 위치는 자동(best location)으로 설정됨. loc 설정을 통해 위치를 옮겨줄수도 있음 
````
