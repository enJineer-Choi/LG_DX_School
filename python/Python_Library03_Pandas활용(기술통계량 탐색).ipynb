#### 실습
- 교육 수준에 따른 평균 연봉

- 교육 수준 카테고리 확인 > unique()
- 불리언 인덱싱을 통해 한개의 카테고리의 평균 연봉 구하기
- 반복문을 활용하여 각각 모든 하력에 대한 평균 연봉 구하기

```python
edu_uni = hr_info['EDU_LEVEL'].unique()
edu_uni
```

- for문에서 list를 받아서 하는 방법 > 사실 groupby 사용하면 됨

```python
for level in edu_uni:
    print(f"{level}: {hr_info[hr_info['EDU_LEVEL'] == level]['SALARY'].mean().round(2)}")
```

- 데이터 그룹화를 이용한 통계분석
  - groupby() : 카테고리별로 그룹을 만들어 연산을 수행
  - 사용법 : 데이터.groupby(기준 컬럼명).연산함수
  - 카테고리로 설정할 데이터가 들어있는 컬럼을 기준컬럼이라고 함

```python
# 기준이 "EDU_LEVEL"
hr_info[['EDU_LEVEL','SALARY']].groupby('EDU_LEVEL').mean()
```

```python
# 교육수준과 성별에 따른 평균연봉
hr_info[['EDU_LEVEL','GENDER','SALARY']].groupby(['EDU_LEVEL','GENDER']).mean() # 기준이 2개 이상이면 리스트 형태로 묶어주어야함
```

```python
# 성별에 따른평균연봉
hr_info[['GENDER','SALARY']].groupby(['GENDER']).mean()
```

- 가설 : 학사학위인 남자가 많을 것이다 > 가설검정

- 그룹화 후에 여러가지 통계값 확인
  - 데이터.groupby(기준).agg(['원하는 통계량'])

```python
# 성별에 따른 연봉의 다양한 통계량 구하기 (평균, 최댓값, 최솟값, 중앙값)
hr_info[['GENDER','SALARY']].groupby(['GENDER']).agg(['mean','max','min','median'])
```

```python
hr_info[['GENDER','SALARY']].groupby(['GENDER']).describe().T
```

#### 상관관계
- 상관분석 : 확률론과 통계학에서 두 변수(컬럼들을 의미)간에 어떤 선형적 관계를 갖고 있는 지를 분석하는 방법
- 상관 관계 : 두 변수는 서로 독립적인 관계
   - 양의 상관관계 : 상관계수가 1에 가까움
   - 음의 상관관계 : 상관계수가 -1에 가까움
   - 상관관계 없음 : 상관계수가 0 에 가까움

- 상관계수 확인 : data.corr()
  - 각 변수간의 선형적 관계를 확인

```python
# 성과평가 점
hr_info[['PERFORMANCE_SCORE','SALARY']].corr()
```

- 성과평가와 연봉은 양의 상관관계를 가짐 (0.609152)

```python
hr_info.corr(numeric_only=True) # numeric_only = True 조건을 줘서 수치형값들만 확인할 수 잇도록
```

- 절대값이 1에 가까울수록 상관관계가 높다
- 근속년수가 높을수록 연봉이 높다

```python
# 데이터정렬
corr_res = hr_info.corr(numeric_only=True)
corr_res['SALARY'].sort_values() # 기본 : 오름차순 정렬
```

```python
corr_res['SALARY'].sort_values(ascending = False) # 내림차순 정렬하려면?
```
