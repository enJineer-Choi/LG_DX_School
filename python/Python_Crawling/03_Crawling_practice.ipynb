### ZDnet 뉴스 데이터 수집 

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = "https://search.zdnet.co.kr/?kwd=%EC%97%94%EB%B9%84%EB%94%94%EC%95%84&area=&pageno=1&term="
res = req.get(url)
soup = bs(res.text,'lxml')
soup
```

```python
title = soup.select("div.assetText>a>h3") # 맞게 불러왔는데 자꾸 데이터 값이 사라져버림;;;
```

```python
title
```

```python
len(soup.select("div.assetText>a>h3"))
```

- 정보를 수집할 때는 명확하게 구분자를 주는 것이 좋다 (반드시 잘 수집되었는지 확인필요!)

```python
link = soup.select("section.news_box div.assetText>a")
link
```

- 수집한 요소를 리스트에 저장(콘텐츠와 링크 정보로 추출후에 저장 필요!)

```python
news_list = []
for i in range(len(title)):
    t = title[i].text
    l = link[i].get('href')
    news_list.append([t,l])
```

```python
pd.DataFrame(news_list,columns = ["제목","링크"])
```

- 현재 첫번째 페이지가 수집된거임! > 다른 페이지들도 가져오려면?

```python
from tqdm import tqdm #반복문의 진행상황을 loading bar로 표기해주는 라이브러리
```

- 논리구조 스스로 이해하고 혼자 해보기!

```python
news_list = []
for i in tqdm(range(1,11)):
    url = f"https://search.zdnet.co.kr/?kwd=%EC%97%94%EB%B9%84%EB%94%94%EC%95%84&area=&pageno={i}&term="
    res = req.get(url)
    soup = bs(res.text,'lxml')
    # 뉴스제목, 링크를 수집
    title = soup.select("div.assetText>a>h3")
    link = soup.select("section.news_box div.assetText>a")
    
    
    for i in range(len(title)):
        t = title[i].text
        l = link[i].get('href')
        news_list.append([t,l])
```

```python
pd.DataFrame(news_list,columns = ['뉴스제목','링크'])
```
