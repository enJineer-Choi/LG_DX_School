### 네이버 뉴스 데이터 수집
- 실습 프로세스
  1. 네이버 뉴스페이지 접속 -> IT/과학
  2. 헤드라인 뉴스 접근(주소받아오기!)
  3. 뉴스의 제목,내용 수집

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = "https://n.news.naver.com/mnews/article/366/0001136987"
```

```python
res = req.get(url)
```

```python
soup = bs(res.text,'lxml')
```

```python
soup
```

```python
title = soup.select('h2#title_area')[0].text #span태그밖에 없어서 이렇게 나올 수 있었음.. 
title
```

```python
soup.select('h2#title_area>span')[0].text # 엄밀하게 하면 이렇게 해야함. 
```

```python
soup.select_one("h2#title_area>span").text #id를 활용하면
```

- span 태그는 구분자가 없어서 부모태그로 올라가서 가져와야함!!
- 리스트 상태에서는 text를 추출할 수 없다! > 반드시 인덱싱을 한 후에 text를 추출해야함

```python
news = soup.select('article#dic_area')[0].text.strip() # strip() 함수를 통해 앞뒤의 \n을 삭제 시키면 됨
news
```

```python
soup.select_one("article#dic_area").text.strip()
```

- id를 활용하면 유일한 값 하나밖에 없으므로 `select_one` 를 사용하면 코드가 더 간결하다!

```python
news_list = []
```

```python
soup = bs(res2.text,'lxml')
```

```python
soup.select("strong.sa_text_strong")[0:5]
```
