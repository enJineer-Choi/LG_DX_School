# Crawling 

- 웹페이지를 그대로 가져와서 데이터를 추출하는 행위
  - 주로 `리뷰데이터` 활용

## 빅데이터 5가지 특정
- 디지털 환경에서 생성되는 수치, 문자, 이미지, 영상데이터 모두를 포함하는 대규모 데이터
  - 규모 (volume)
  - 속도 (velocity)
  - 다양성 (variety)
  - 정확성 (varacity)
  - 가치 (value)

### 수집데이터의 형태
- 정형데이터
  - 데이터베이스(DB), 스프레드시트, csv
- 반정형데이터
  - html,xml
- 비정형데이터
  - 사진,음성,음악,영상

### Client와 Server란?
- client : 클릭한 페이지를 요청하는 pc
  - 네트워크로 연결된 서버로부터 정보를 제공받는 컴퓨터
- server : 클릭된 페이지를 제공하는 pc
  - 클라이언트에게 네트워크를 통해 서비스를 제공하는 컴퓨터
- client -> server : Request
- server -> client : Response
- 서버가 전송해준 응답을 디자인을 입혀주는 역할 : 브라우저

### URL이란?
- Uniform Resource Locator
- 웹문서의 수많은 서비스를 제공하는 서버들에 잇는 파일의 위치를 표시하는 표준 주소

### Request 라이브러리?
- 접근할 웹페이지의 데이터를 요청/응답 받기 위한 라이브러리
