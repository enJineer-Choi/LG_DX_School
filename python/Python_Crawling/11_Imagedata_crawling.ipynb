````python
#  이미지의 src 수집하여 img_list에 담기
````

````python
#1. 드라이버에게 p태그 위치 알려주기
p = driver.find_elements(By.CSS_SELECTOR,'div.bx-txt>h3>p')
````

````python
poke_name = [n.text for n in name]
````
