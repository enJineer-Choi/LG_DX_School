##### 문자열의 인덱싱과 슬라이싱 
- indexing : 무엇인가를 가리킨다는 의미, 인덱스를 활용해서 문자 가져오기
- slicing : 무엇인가를 잘라낸다는 의미, 인덱스를 활용해서 원하는 구간 범위만큼의 데이터를 가져오기

```python
s = "My name is SJ"
print(s[12])
print(s[-5])
```

```python
print(s[-5:-3])
print(s[8:10])
```

```python
print(s[3:7])
```

```python
day = '2025년 12월 31일의 날씨는 흐림입니다.'
print("날짜 : ", day[0:13])
print("날씨 : ", day[-6:-4])
```

```python
day[:13]
```

```python
day[-6:-4]
```

```python
day[-6:]
```

##### 문자열 포맷팅(formatting)
- 문자열에 특정한 데이터를 포함시키고자 할 때 사용
- 1. % 기호 formatting : 모든 프로그래밍 언어에서 통용되는 방법
  2. format 함수 : 파이썬에서만 사용 가능 
  3. f문자열 formatting : 파이썬에서만 사용 가능

1. % 기호 formatting



```python
y = '2025'
m = 12
d = '31'
w = '흐림'

print('오늘은 %s년 입니다.'%y)
print('오늘은 %s년 %d월 %s일 입니다.'%(y,m,d))
```

2. format 함수 : 파이썬에서만 사용 가능 

```python
y = '2025'
m = 12
d = '31'

s = '오늘은 {}년 {}월 {}일 입니다.'.format(y,m,d)
print(s)
```

3. f문자열 formatting : 파이썬에서만 사용 가능

```python
y = '2025'
m = 12
d = '31'

print(f"{y}년 {m}월 {d}일 입니다")
```
