- 모듈(라이브러리) 내에 자주 사용하는 함수만 불러와서 사용
> 라이브러리에 너무 많은 기능과 데이터가 포함되서 무거운 경우에 활용
```
from Module(library 명) import add_num(library 내 함수)
from Module(library 명) import sub_num(library 내 함수)
```
    - 모듈 내의 특정함수만 불러온 경우 다른 함수는 사용 불가!

````python
from Module import add_num
````

````python
add_num(20,30)
````

````python
from Module import sub_num
sub_num(30,20)
````

- 라이브러리 (library) > 도서관
- 라이브러리 (library) 는 모듈(module)이라고도 불리며, 자주쓰는 함수, 변수들을 모아놓은 파이썬 파일.
- 자주 사용하는 기능을 모듈화, 패키지화 하여 만들어 둔 것.


- 표준 라이브러리
  - Python에서 기본으로 제공하는 라이브러리
- 외부 라이브러리
  - 개발자가 필요에 의해 개발한 패키지와 모듈의 집합
