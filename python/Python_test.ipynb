list1 = [1,50,27,19,37,33,5,3,48]  다음과 같은 리스트에서 홀수와 짝수의 개수를 확인하는 문제

````python
list1 = [1,50,27,19,37,33,5,3,48]

even_number = 0
odd_number = 0

for num in list1:
    if num%2 == 0:
        even_number += 1
    else:
        odd_number += 1
print(f"짝수 : {even_number}, 홀수 : {odd_number}")
````

list2 = [4,5,2,1,99,15,2,7,27] 다음 리스트에서 가장 큰 수 찾는 문제

````python
list2 = [4,5,2,1,99,15,2,7,27]
big_number =list2[0]

for num in list2:
    if num > big_number:
        big_number = num
print(f"가장 큰 수는 {big_number}입니다")
````
