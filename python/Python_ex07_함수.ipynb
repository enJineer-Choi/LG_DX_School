- 두개의 정수와 한개의 연산기호(+,-)를 매개변수로 받아 연산기호에 맞게 연산하여 결과값을 반환

````python
def cal(n1,n2,c):
    if c == "+":
        return n1+n2
    elif c == "-":
        return n1-n2
````

- 두개의 정수를 매개변수로 받아, 두숫자 중 10에 더 가까운 수를 반환하는 close10 함수 만들기
  - 만약 두 숫자 모두 10과의 차이가 같다면 0을 반환

````python
def close10(n1,n2):
    if abs(10-n1) < abs(10-n2):
        return n1
    elif abs(10-n1) > abs(10-n2):
        return n2
    else:
        return 0
````

````python
close10(4,6)
````

````python
close10(9,9)
````

````python
close10(9,8)
````

````python
close10(9.999,9.888)
````

- 두개의 숫자를 (start,end)를 매개변수로 받아서 start부터 end까지의 숫자를 list로 만들어 반환


````python
def rangeList(start,end):
    l = []
    for i in range(start,end+1):
        l.append(i)
    return l
````

````python
rangeList(1,10)
````

````python
def rangeList_1(start,end):
    return print([i for i in range(start,end+1)])
````

````python
rangeList_1(10,30)
````

- 선생님의 풀이

````python
def rangeList(start,end):
    return list(range(start,end+1))
````

````python
# 함수정의
def cal(n1,n2,op):
    """함수 설명 : 기호를 입력받아 덧셈과 뺄셈을 계산하는 함수""" # 
    if op == "+":
        return n1+n2
    elif op == "-":
        return n1-n2
    else:
        return "기호를 잘못 입력하셨습니다"
````

````python
# 함수호출
num1 = int(input("첫번째 숫자를 입력하세요 >> "))
num2 = int(input("두번째 숫자를 입력하세요 >> "))
oper = input("연산자 입력(+,-) >> ")
result = cal(num1,num2,oper)
print(f"결과 : {result}")
cal()
````

- 함수 - 매개변수 기본값 설정> 매개변수에 미리 값을 부여

````python
# 숫자 데이터에 제곱을 구해서 되돌려주는 함수
def power_num(num,power = 2 #default 설정):
    return num**power
````

````python
# 기본값으로 계산 > power 입력하지 않아도 OK
power_num(3)
````

````python
# 세제곱으로 변경할 때는 ,power에 원하는 숫자 입력
power_num(3,3),power_num(3,power=3)
````

````python
def all_sum(*args):
    result = 0
    for num in args: # args(튜플)의 모든 데이터를 num에 담아 연산
        result += num
    return result
````
