##### Series 연산

````python
# dictionary(key : index, value : value)로 데이터 생성
data = {'꼬부기' : 1256, '이브이' : 2542, '아르세우스' : 9585, '치코리타' : 100}
pokemon = pd.Series(data)
pokemon
````

````python
data2 = {'꼬부기' : 1500, '이브이' : 3000,'마자용': 200,'아르세우스' : 10000}
pokemon_up = pd.Series(data2)
pokemon_up
````

````python
# index를 이름으로 지정했으므로, 이름별로 연산 진행
result = pokemon_up - pokemon
result
````

- index 가 동일한 데이터끼리 연산
- 결측치와의 연산 결과는 결측치 
  - 데이터 분석시 결측치가 존재하면 결측치 제거 후 진행

- 결측치 처리 방법
  - 결측치 확인 : ```변수명.isnull()``` > 결과는 논리형
  - 비어있지 않은 값만 확인 : ```변수명.notnull()```
  - 결측치 찾아 특정값으로 변경(대입) : ```변수명.fillna(원하는 값)``` > result의 결과가 직접 바뀌지 않음!
  - 삭제 : ```del 변수명``` > 삭제 후 바로 적용

````python
result.isnull()
````

````python
# 비어있는 값들만 출력 -> 조건 (boolean indexing)
result[result.isnull() == True]
````

````python
# 비어있지 않은 값들
result.notnull()
````

````python
result[result.notnull()==True]
````

````python
# 데이터 업데이트 > 마자용 20 결과를 업데이트(업데이트의 의미 : 새로운 값을 대입)
result['마자용'] = 20
result
````

````python
# 데이터 추가 > 미뇽 150 추가 결과 
result['미뇽'] = 150
result
````

- 업데이트 : 원래있는 인덱스 값에 새로운 값 대입
- 추가 : 새로운 인덱스에 값 대입

````python
# 비어있는 Nan 에 동일값 일괄 대입
result = result.fillna(0)
result
````

````python
# 데이터 삭제 
del result['치코리타']
result
````

#### DataFrame
- DataFrame : 행과 열을 가지는 2차원 데이터

````python
team = pd.DataFrame([
            ['김','ESFJ'],
            ['이','INTP'],
            ['박','ISFP'],
            ['최','INTP'],
            ['정','ESTJ'],
            ['장','INFJ']],
            columns = ['이름','MBTI'],
            index = ['001','002','003','004','005','006'])
team
````

````python
# 전치 : transpose(), 변수명.T 도 가능!
team.transpose()
````

````python
team.T
````

````python
# Dictionary 활용하여 Dataframe 생성하기
# key > columns, value > value
team_dict = {
    '이름': ['구상원', '김소은', '최보선', '신현경', '신윤지', '장지석'],
    'MBTI': ['ESFJ', 'INTP', 'ISFP', 'INTP', 'ESTJ', 'INFJ']
}

team = pd.DataFrame(
    team_dict,
    index=['001', '002', '003', '004', '005', '006']
)

team
````

### 데이터 색인(심화) : 인덱서(loc,iloc)
- 행과 열의 범위를 설정하여 데이터를 추출
- ```df.loc[행범위,열범위]``` : 컬럼명, 인덱스명 통해서 추출(이름을 직접 사용해야함!)
- ```df.iloc[행범위,열범위]``` : 인덱스 번호를 통해서 추출
- loc,iloc 인덱싱을 통해 `행 인덱싱`이 가능해짐

````python
hr_info.loc['E00206':'E00164','ENAME':'AGE']
````

````python
hr_info.iloc[0:3,0:5]
````

````python
hr_info.iloc[:3,:5] #처음부터면 0 생략가능!
````

````python
# 모든 사람의 age ~ salary 까지 > iloc 사용
hr_info.iloc[:,4:9]
````

````python
# 모든 사람의 age ~ salary 까지 > iloc 사용
hr_info.loc[:,"AGE":"SALARY"]
````

```hr_info.loc["AGE":"SALARY"]``` > 전체행이라고 "행" 을 생략하면 안됨.

````python
hr_info.iloc[0:7,1:7]
````

````python
hr_info.iloc[:3] #comma를 작성해주지 않으면 행단위로 인덱싱이 가능
````

#### 데이터 삽입, 삭제, 추가...

- columns 추가
  - df.insert(loc = 인덱스, column = 추가할 컬럼명, value = 데이터)
  - loc 속성에 작성한 인덱스 자리에 삽입 -> 그 이후 데이터는 뒤로 밀리게 됨
  - df['새로운 컬럼명'] = 데이터

````python
#'MONTHLY_PAY' 추가
hr_info.insert(loc = 9, column = 'MONTHLY_PAY', value = hr_info['SALARY']/12)
hr_info
````

````python
#'MONTHLY_PAY' 추가 > 이렇게 추가해도 됨!
hr_info['DAILY_PAY'] = hr_info['MONTHLY_PAY']/30
hr_info
````

- 데이터 수정
  - df['존재하는 컬럼명'] = 업데이트할 데이터

````python
hr_info['DAILY_PAY'] = (hr_info['MONTHLY_PAY']/30).round(2)
hr_info
````

- 행(row) 추가
  - df.loc['추가하고 싶은 인덱스'] = [데이터]
  - 추가할 데이터에 들어가는 특성의 개수(column 개수)를 맞춰줘야 한다.
  - 사실 거의 하지 않음. (새로운 데이터가 들어올 때)
