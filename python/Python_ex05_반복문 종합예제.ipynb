```python
print("~ chef's kitchen ")
menu = ["Burger","Pizza","Pasta","Salad","Sushi"]
baggage = []

while True :
    choice = int(input("[1] [2] [3]()\n  : "))
    if choice == 3 :
        quit = input(" ?  'q'  : ")
        if quit == "q":
            print(" !")
            break
    elif choice == 1: 
        print("")
        for m in menu:
            print(m)
        m_count = int(input("    : "))
        for i in range(m_count):
            m_choice = input("  : ")
            baggage.append(m_choice)
        print(f" : {baggage}")
        print()
    elif choice == 2:
        del_menu = input("   : ")
        baggage.remove(del_menu)
        print(f"{del_menu}()  .")
        print(f" : {baggage}")
        print()
```
