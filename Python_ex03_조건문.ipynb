### 조건문
 - 상황에 따라 실행 흐름을 다르게 하는 문법

### if문 
```
if 조건식 : 
    실행문장
    실행문장
```

````python
if True :
    print('실행문장 1 실행')
print('조건문 종료')
````

````python
if False :
    print('실행문장 1 실행')
print('조건문 종료')
````

````python
if True :
    print('실행문장 1 실행')
        print('실행문장 2 실행')
print('조건문 종료')
````

### if-elif 조건문
```
if 조건식1 : 
    실행문장1
elif 조건식2 : 
    실행문장2
else:
    실행문장3
```
- 조건식이 `True` 일 경우 `실행문장1` 실행
- 조건식이 `False`일 경우 `다음 조건식 확인`

````python
if True : 
    print('실행문장1')
elif True : 
    print('실행문장2')
elif True : 
    print('실행문장3')
else:
    print('실행문장4')

print('if문 종료')
````

````python
num1 = int(input("첫번째 숫자 입력 >>"))
num2 = int(input("두번째 숫자 입력 >>"))

if num1 > num2:
    print("첫번째 숫자가 더 큽니다")
elif num1 == num2 : 
    print("두 수가 똑같습니다")
else:
    print("두번째 숫자가 더 큽니다")
````
