- 리스트와 튜플의 공통점 & 차이점



````python
lst = ['딸기', '바나나','수박','체리','포도']
print("*"*40)
word = input("검색할 문자를 입력하세요 >>")

if word in lst:
    print(f"{word}는 리스트의 {lst.index(word)}번째 인덱스에 들어있습니다.")
    print(f"{word}는 리스트의 {lst.index(word)+1}번에 들어있습니다.")
else:
    print(f"{word}는 리스트에 들어있지 않습니다.")
print("*"*40)
````

````python
s = "Hi, My name is Seokjin"

word = input("검색할 문자를 입력하세요 : ")

if word in s:
    print(f"'{word}'는 {s.count(word)} 번 들어 있습니다.")
else:
    print(f"'{word}'는 들어있지 않습니다.")
````

````python
# 소문자로 바꿔서 찾기 (의미상 유사하도록)
s = "Hi, My name is Seokjin"

word = input("검색할 문자를 입력하세요 : ")

s_lower = s.lower() # 문자열 소문자로 변경

if word in s_lower:
    print(f"'{word}'는 {s_lower.count(word)} 번 들어 있습니다.")
else:
    print(f"'{word}'는 들어있지 않습니다.")
````

````python
# 리스트 버전
lst = []

first = input("첫 번째 요리는? ")
sec = input("두 번째 요리는? ")
thr = input("세 번째 요리는? ")

lst.append(first)
lst.append(sec)
lst.append(thr)


print(f"다시 확인할게요. 주문하신 메뉴는 {lst[0]},{lst[1]},{lst[2]}입니다.")

new = input("마지막 요리를 변경하시겠어요?? 무엇으로 바꿀까요? ")

# lst.remove(thr)
# lst.append(new)

lst[-1] = new # 마지막 값 자리에 그냥 new를 넣어도 된다.

print(f"다시 확인할게요. 주문하신 메뉴는 {lst[0]},{lst[1]},{lst[2]}입니다.")
````

````python
lst = []

for i in range(1,4):
    lst.append(input(f"{i} 번째 요리는? "))
    

print(f"다시 확인할게요. 주문하신 메뉴는 {lst[0]},{lst[1]},{lst[2]}입니다.")

new = input("마지막 요리를 변경하시겠어요?? 무엇으로 바꿀까요? ")

del lst[-1]
lst.append(new)

# lst[-1] = new

print(f"다시 확인할게요. 주문하신 메뉴는 {lst[0]},{lst[1]},{lst[2]}입니다.")
````

````python
lst = []

for i in range(1,4):
    lst.append(input(f"{i} 번째 요리는? "))
    
for i in range(3):
    print(f"다시 확인할게요. 주문하신 메뉴는 {lst[i]}입니다.")

new = input("마지막 요리를 변경하시겠어요?? 무엇으로 바꿀까요? ")

del lst[-1]
lst.append(new)

# lst[-1] = new

print(f"다시 확인할게요. 주문하신 메뉴는 {lst[0]},{lst[1]},{lst[2]}입니다.")
````
